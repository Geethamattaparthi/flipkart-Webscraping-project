```python
import pandas  as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import requests
import re
from bs4 import BeautifulSoup
```

##### Webscraping from flipkart website

```python
brand=[]
price=[]
discount=[]
color=[]
rom=[]
rating=[]
front_cam=[]
back_cam=[]
processor=[]


for i in range(1,61):
    url=f"https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_7_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_7_na_na_ps&as-pos=1&as-type=HISTORY&suggestionId=mobiles%7CMobiles&requestId=d3382d16-c54a-4313-bdbe-040a559ac9d9&page={i}"
    #print(url)
    page=requests.get(url)
    soup=BeautifulSoup(page.text)
    
#brand    
    for i in soup.find_all("div",class_="_4rR01T"):
            print(i.text)
            regex=re.findall("^[a-zA-Z]+",i.text)
            print(regex)
            if regex:
                brand.append(regex[0])
            else:
                brand.append(np.nan)
                
#price                
    for i in soup.find_all('div',class_='_2kHMtA'):
        text=i.find('div',class_='_30jeq3 _1_WHN1')
        if text:
            price.append(text.text)
        else:
            price.append(np.nan)          
                
#discount                
    for i in soup.find_all("div",class_="_2kHMtA"):
        a=i.find("div",class_="_3Ay6Sb")      
        if a:
            regex = re.findall('^\d+',a.text)
            discount.append(regex[0])
        else:
            discount.append(np.nan)
#color            
    for i in soup.find_all("div",class_="_4rR01T"):
        print(i.text)
        regex=re.findall("\((\w+\s?\w+),",i.text)
        print(regex)
        if regex:
            color.append(regex[0])
        else:
            color.append(np.nan)
#storage            
    #for i in soup.find_all("div",class_="_4rR01T"):
        #print(i.text)
        #regex=re.findall("(\d+)\sGB",i.text)
        #print(regex)
        #if regex:
            #storage.append(regex[0])
        #else:
            #storage.append(np.nan)
#display size            
    
    #for i in soup.find_all("li",class_="rgWa7D"):
        #text=i.text
        #regex=re.findall("(\d+\.\d+) inch",text)
        #if regex:
            #display_size.append(regex[0])
        #else:
            #display_size.append(np.nan)
        
#RAM            
    container=soup.find_all("div",class_="_3pLy-c row")        
    for i in container:
        c=i.find("li",class_="rgWa7D")
        print(c.text)
        regex=re.findall("(\d+)\sGB\sRAM",c.text)
        if regex:
            rom.append(regex[0])
        else:
            rom.append(np.nan)
#rating        
    for i in soup.find_all('div',class_='_2kHMtA'):
        b=i.find("div",class_="_3LWZlK")
        if b:      
            rating.append(b.text)
        else:
            rating.append(np.nan)
            
#f_cam
    for i in soup.find_all('div',class_='_2kHMtA'):
        text=i.text
        regex = re.findall('(\d+)MP Front Camera',text)
        if regex:
            front_cam.append(regex[0])
        else:
            front_cam.append(np.nan)
#b_cam           
    for i in soup.find_all('div',class_='_2kHMtA'):
        text=i.text
        regex = re.findall('(\d+MP \+\s?\d+MP)',text)
        if regex:
            back_cam.append(regex[0])
        else:
            back_cam.append(np.nan)
            
            
    #for i in container:
        #a=i.find("li",class_="rgWa7D")
        #print(a.text)
        #regex=re.findall("(\d+)\smAh",c.text)
        #if regex:
            #battery.append(regex[0])
        #else:
            #battery.append(np.nan)
            
            
    #for processor_tag in soup.find_all("li", class_="rgWa7D"):
        ##print(processor_tag.text)
        #processor_regex = re.findall(r'\bDimensity\s+\d+\b|\bSnapdragon\s+[a-zA-Z0-9]+\b|\bA\d+\s+Bionic\b', processor_tag.text)
        #print(processor_regex)
        #if processor_regex:
            #processor.append(processor_regex[0])
        #else:
            #processor.append(np.nan)

                
                       
                
```
